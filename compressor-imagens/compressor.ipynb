```python
from PIL import Image

def comprimeImagemRle(aImagem):
    aImagemComprimida = []
    
    for aLinha in aImagem:
        aLinhaComprimida = []
        iPixelAtual = aLinha[0]
        iContador = 1
        
        for iPixel in aLinha[1:]:
            if iPixel == iPixelAtual:
                iContador += 1
            else:
                aLinhaComprimida.extend([iPixelAtual, iContador])
                iPixelAtual = iPixel
                iContador = 1
        
        aLinhaComprimida.extend([iPixelAtual, iContador])
        aImagemComprimida.append(aLinhaComprimida)
    
    return aImagemComprimida

def descomprimeImagemRle(aImagemComprimida):
    aImagemDescomprimida = []
    
    for aLinha in aImagemComprimida:
        aLinhaDescomprimida = []
        
        for i in range(0, len(aLinha), 2):
            iPixel = aLinha[i]
            iContador = aLinha[i + 1]
            aLinhaDescomprimida.extend([iPixel] * iContador)
        
        aImagemDescomprimida.append(aLinhaDescomprimida)
    
    return aImagemDescomprimida

def carregaImagem(sArquivo):
    oImagem = Image.open(sArquivo)
    return list(oImagem.getdata())

def getTamanhoImagem(sArquivo):
    oImagem = Image.open(sArquivo)
    return oImagem.size

def salvaImagem(aImagem, iLargura, iAltura, sArquivo):
    aImagemFlat = [iPixel for aLinha in aImagem for iPixel in aLinha]

    oImagem = Image.new("RGB", (iLargura, iAltura))
    oImagem.putdata(aImagemFlat)
    oImagem.save(sArquivo)

def comprimirImagemPpm(sArquivo, sSalvar):
    aImagemOriginal = carregaImagem(sArquivo)
    aTamanhoImagem = getTamanhoImagem(sArquivo)
    iLargura, iAltura = aTamanhoImagem
    aImagemOriginal = [list(aImagemOriginal[i:i + iLargura]) for i in range(0, len(aImagemOriginal), iLargura)]

    aImagemComprimida = comprimeImagemRle(aImagemOriginal)
    aImagemDescomprimida = descomprimeImagemRle(aImagemComprimida)
    salvaImagem(aImagemDescomprimida, iLargura, iAltura, sSalvar)

comprimirImagemPpm("imagemHomogenea.ppm", "comprimidaHomogenea.ppm")
comprimirImagemPpm("imagemHeterogenea.ppm", "comprimidaHeterogenea.ppm")
```
